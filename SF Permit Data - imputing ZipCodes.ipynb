```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Permit Data

```python
sf_permits = pd.read_csv('Building_Permits.csv')
sf_permits.head()
```

How much data is missing?

```python
missing_perm_counts = sf_permits.isnull().sum()
missing_perm_counts.head(10)
```

Express as percent of total # entries:

```python
total_perm_rows = len(sf_permits)
missing_perm_df = pd.DataFrame({'Missing Data Counts':missing_perm_counts})
missing_perm_df['Missing Data %'] = missing_perm_df['Missing Data Counts'] / total_perm_rows *100
missing_perm_df.head(20)
```

### Which data doesn't exist, and which simply wasn't recorded?

Look at 'Street Number Suffix':

```python
sf_permits[sf_permits['Street Number Suffix'].notnull()]
```

Appears to be an Apartment letter. Do not expect every entry to have one.

What about Zipcode?

```python
missing_perm_df.loc['Zipcode']
```

```python
sf_permits[sf_permits['Zipcode'].isnull()]
```

Missing Zipcode entries are likely so because they weren't recorded.

Maybe we can guess what they are based on Block ID?

```python
sf_permits[sf_permits['Block'] == '0552']
```

```python
sf_permits[sf_permits['Block'] == '1040']['Zipcode'].unique()
```

Yup, all permits with a given Block ID have the same Zipcode.

#### Create dictionary of block-zipcode pairs:

```python
block_zip_dict = {}
for b in sf_permits[sf_permits['Zipcode'].isnull()]['Block'].unique():
        l = sf_permits[sf_permits['Block'] == b]['Zipcode'].unique()
        if len(l) > 1:
            z = l[np.isnan(l) == False][0]
        else:
            z = None
        block_zip_dict[b] = z
```

```python
block_zip_dict
```

Apply to df:

```python
sf_permits.loc[sf_permits['Zipcode'].isnull(), 'Zipcode'] = sf_permits.loc[sf_permits['Zipcode'].isnull(), 'Block'].map(block_zip_dict)
```
